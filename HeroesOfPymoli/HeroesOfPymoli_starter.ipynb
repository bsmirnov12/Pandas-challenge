```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count


* Display the total number of players

```python
unique_players = len(purchase_data['SN'].unique())
total_players_df = pd.DataFrame([{'Total Players': unique_players}])

total_players_df
```

## Purchasing Analysis (Total)


* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Number of purchases
purchases_num = len(purchase_data)

# Total revenue
total_revenue = sum(purchase_data['Price'])

# Average price
average_price = total_revenue / purchases_num

# Number of unique items
unique_items = len(purchase_data['Item ID'].unique())

# Formatting & Displaying Report
purchase_analysis_total = pd.DataFrame([{
    'Number of Unique Items': unique_items,
    'Average Price': "${:.2f}".format(average_price),
    'Number of Purchases': purchases_num,
    'Total Revenue': "${:,.2f}".format(total_revenue)
}])

purchase_analysis_total
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed

```python
# 1. Get unique players and their gender info
players_grp = purchase_data.groupby('SN')
players_gender_s = players_grp['Gender'].first()

# 2. Calculate Gender Stats
gender_demographics_df = pd.DataFrame(players_gender_s.value_counts())
gender_demographics_df.columns = ['Total Count']
gender_demographics_df['Percentage of Players'] = (gender_demographics_df['Total Count'] / unique_players).map("{:.2%}".format)

# 3. Display
gender_demographics_df
```

## Purchasing Analysis (Gender)


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# 1. Gathering purchase stats for gender groups
genders_grp = purchase_data.groupby('Gender')

# 2. Calculating basic stats
purchase_count_s = genders_grp['Purchase ID'].count()
avg_purchase_price_s = genders_grp['Price'].mean()
total_purchase_val_s = genders_grp['Price'].sum()

# 3. Calculating Average Total Purchase per Player
gender_player_grp = purchase_data.groupby(['Gender', 'SN'])
avg_total_purchase_s = gender_player_grp['Price'].sum().mean(level=0)

# 4. Formatting & Displaying Report
purchase_analysis_gender = pd.DataFrame({
    'Purchase Count': purchase_count_s,
    'Average Purchase Price': avg_purchase_price_s.map("${:.2f}".format),
    'Total Purchase Value': total_purchase_val_s.map("${:,.2f}".format),
    'Avg Total Purchase per Person': avg_total_purchase_s.map("${:,.2f}".format)
})

purchase_analysis_gender
```

## Age Demographics


* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table

```python
# 1. Establishing bins for ages
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 120]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# 2. Categorizing
players_age_df = pd.DataFrame(players_grp['Age'].first())
players_age_df['Age Group'] = pd.cut(players_age_df['Age'], age_bins, labels=age_labels)

# 3. Calculating stats
age_demographics_df = pd.DataFrame(players_age_df['Age Group'].value_counts(sort=False))
age_demographics_df.columns = ['Total Count']
age_demographics_df['Percentage of Players'] = (age_demographics_df['Total Count'] / unique_players).map("{:.2%}".format)

age_demographics_df
```

## Purchasing Analysis (Age)


* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# 1. Categorizing Purchase Data
purchase_data['Age Group'] = pd.cut(purchase_data['Age'], age_bins, labels=age_labels)
purchase_age_grp = purchase_data.groupby('Age Group')

# 2. Calculating basic stats
purchase_count_by_age_s = purchase_age_grp['Purchase ID'].count()
avg_purchase_price_by_age_s = purchase_age_grp['Price'].mean()
total_val_by_age_s = purchase_age_grp['Price'].sum()

# 3. Calculating Average Total Purchase per Player by Age Group
age_player_grp = purchase_data.groupby(['Age Group', 'SN'])
avg_total_purchase_s = age_player_grp['Price'].sum().mean(level=0)

# 4. Formatting & Displaying Report
purchase_analysis_age_group = pd.DataFrame({
    'Purchase Count': purchase_count_by_age_s,
    'Average Purchase Price': avg_purchase_price_by_age_s.map("${:.2f}".format),
    'Total Purchase Value': total_val_by_age_s.map("${:,.2f}".format),
    'Avg Total Purchase per Person': avg_total_purchase_s.map("${:,.2f}".format)
})

purchase_analysis_age_group
```

## Most Popular Items


* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame

```python
# 1. Get Unique Items and their Descriptive Info
items_grp = purchase_data.groupby(['Item ID', 'Item Name'])
items_prices_s = items_grp['Price'].first()

# 2. Calculate Basic Stats
items_purchase_count_s = items_grp['Purchase ID'].count()
items_total_val_s = items_grp['Price'].sum()

# 3. Semi-ready for display... Items Stats DataFrame
items_stats_df = pd.DataFrame({
    'Purchase Count': items_purchase_count_s,
    'Item Price': items_prices_s.map("${:.2f}".format),
    'Total Purchase Value': items_total_val_s.map("${:.2f}".format),
    '$': items_total_val_s # temporary column for sorting
})

# 3. Formatting & Displaying Report
# Logically speaking, if two items have the same purchase count, then the one with higher price is more popular then another
# Thus introduced additional temporary column '$' with numeric value suitable for sorting items with the same purchase count
most_popular_items_df = items_stats_df.sort_values(by=['Purchase Count', '$'], ascending=False).iloc[:5, 0:3]

most_popular_items_df
```

## Most Profitable Items

```python
# 3. Formatting & Displaying Report
most_profitable_items_df = items_stats_df.sort_values(by=['$'], ascending=False).iloc[:5, 0:3]

most_profitable_items_df
```

## Item Price Rank

* It interesting to notice, that 4 out of 5 items from Top 5 of the most profitable and the most popular items coincide.
* It is also interesting to look into relative pricing of these items using rank function - it produces price ranks relative to the most expensive item (rank = 1).

```python
items_stats_df['Price Rank'] = items_stats_df['Item Price'].rank(pct=True)
items_stats_df.loc[
    [items_stats_df['Price Rank'].idxmax(), items_stats_df['Price Rank'].idxmin()],
    ['Purchase Count', 'Item Price', 'Price Rank', 'Total Purchase Value']
]

most_popular_items_df = items_stats_df.sort_values(by=['Purchase Count', '$'], ascending=False).iloc[:5, [0, 1, 4, 2]]
most_profitable_items_df = items_stats_df.sort_values(by=['$'], ascending=False).iloc[:5, [0, 1, 4, 2]]

popular_and_profitable = pd.merge(
    most_popular_items_df, most_profitable_items_df['Price Rank'],
    how='inner', left_index=True, right_index=True).iloc[:, 0:4].sort_values(by='Price Rank_x', ascending=False)
popular_and_profitable.rename(columns={'Price Rank_x': 'Price Rank'}, inplace=True)
popular_and_profitable
```

* The table above demonstrates, that the most profitable and popular items are in the top 20% of the most expensive ones. That means, that players are willing to buy expensive items, even if they have a much wider choice of less expensive ones. This, of cause, can be explained by the value of those items in the game (usfulness per $).
* It might be tempting to increase the price of those popular items, but that is a very unpopular move among the players, and it might actually cause decrease in revenues
* There might be a way of stimulating the players to buy more of those items, for example, by eliminating some similar, but less "powerful" and cheaper items from the game store.
* Anothe aproach is to carefully exemine game features of those popular items and devise new similar ones carefully balancing their in game "power" and price, thus attracting players who are hesitating to pay for an expensive powerful item, but are ok with paying for less powerful, but less expensive one.
